```python
from decimal import *
getcontext().prec = 6000
```

```python
def get_reverse_dec(dec):
    dec_part = str(dec)
    rev = dec_part[::-1]
    rev = rev[1:-2]
    return rev
```

```python
def has_recurring(rev_str):
    """ input is a rev_str
    
    returns number of recurring part in decimal
    """
    
    # for every recurring cycle length 1->50
    for rec_l in range(1, 1000):
        
        # get original string to compare parts
        compare_str = rev_str[0:rec_l]
        
        # try 5 times
        for i in range(1, 6):
            
            # get next part and compare
            if compare_str != rev_str[rec_l * i:rec_l * (i+1)]:
                
                # if not the same, break loop for this length
                break
        else:    
            # if all were the same, return length
            return rec_l
    
    return 0
```

```python
max_rec = -1
answer = -1

for i in range(1, 1000, 1):
    numerator = Decimal(1)
    denominator = Decimal(i)
    rev_str = get_reverse_dec(numerator/denominator)
    
    if len(rev_str) >= 799:
        rec_cycle = has_recurring(rev_str)
        
        if rec_cycle > max_rec:
            max_rec = rec_cycle
            answer = denominator
            
        print(f"n: {i}, str: {rev_str[:20]} has recurring: {rec_cycle}")

print(f"Answer: {answer} has recurring cycle: {max_rec}")
```
